```python
import os 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
zeno_path = r'C:\Users\mmccu\Box\MM_Personal\5_Projects\BoveLab\3_Data_and_Code\test_velocity_01302025\zeno_all_height_proxy_outputs\depth_proxy_all.csv'
zeno_df = pd.read_csv(zeno_path, index_col = 0)
 
zeno_output_path = r'C:\Users\mmccu\Box\MM_Personal\5_Projects\BoveLab\3_Data_and_Code\test_velocity_01302025\zeno_all_height_proxy_outputs\histogram'
zeno_df.head()
```

```python
# PWS - all 
zeno_pws_df = zeno_df.loc[zeno_df['filename'].str.contains('PWS_1')]
plt.hist(zeno_pws_df['delta_pix_h_rel_50%'])
plt.ylabel('pws_delta_pix_h_rel_50%_all')
plt.savefig(os.path.join(zeno_output_path, 'pws_delta_pix_h_rel_50%_hist_all.png'))
plt.close()
```

```python
# FW all 
zeno_fw_df = zeno_df.loc[zeno_df['filename'].str.contains('FW_1')]
plt.hist(zeno_fw_df['delta_pix_h_rel_50%'])
plt.ylabel('delta_pix_h_rel_50%_all')
plt.savefig(os.path.join(zeno_output_path, 'fw_delta_pix_h_rel_50%_hist_all.png'))
plt.close()
```

```python
plt.hist(zeno_fw_df.loc[zeno_fw_df['delta_pix_h_rel_50%'] <= 1]['delta_pix_h_rel_50%'])
plt.ylabel('fw_delta_pix_h_rel_50% < 1')
plt.savefig(os.path.join(zeno_output_path, 'fw_delta_pix_h_rel_50%_hist_under1.png'))
plt.show()
```
