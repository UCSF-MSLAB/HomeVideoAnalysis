```python
import os
import cv2
import pandas as pd
import mediapipe as mp
from mediapipe.framework.formats import landmark_pb2
```

```python
# load csv 
position_data_parent = r'C:\Users\mmccu\Box\MM_Personal\5_Projects\BoveLab\3_Data_and_Code\gait_bw_zeno_outputs_006'
position_data_file = r'subset_for_memory_c\BW-0340\2024_07_17\000_raw_pose_data\gait_vertical_FW_1_mediapipe.csv'
full_position_path = os.path.join(position_data_parent, position_data_file)
print(full_position_path)
df = pd.read_csv(full_position_path)
```

```python
# set up mediapipe drawing 
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
# Load video
video_filepath = r"C:\Users\mmccu\Box\MM_Personal\5_Projects\BoveLab\3_Data_and_Code\2024_10_Megan_BW_Zeno\BW-0002\2022_09_12\gait_vertical_FW_1.mp4"

cap = cv2.VideoCapture(video_filepath) # video file path 
fps = int(cap.get(cv2.CAP_PROP_FPS)) # gets frames per second 
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # gets image width 
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # gets image height 

if not cap.isOpened():
    print("Error: Could not open video file.")
else:
    print("Video file opened successfully.")
```

```python
# set file path to save video in same folder as raw data 
# use position data file path and data file to reconstruct ifle path 

root, ext = os.path.splitext(position_data_file)
print(root)
print(ext)
video_file = f'{root}_mpVideo.mp4"'
print(video_file)

video_out_path = os.path.join(position_data_parent, video_file) 

# check ful path exists 
before, after = os.path.split(video_out_path)

# check path exits 
if os.path.exists(before):
    print(f"Exists: {before}")
else:
    print(f"Does not exist: {before}")
```

```python
# create video writer object to save video with landmarks 
out = cv2.VideoWriter('test.mp4',  # file name 
                      cv2.VideoWriter_fourcc(*'mp4v'), # sets video codec - codec compabtible with most mp4 files 
                      fps, (width, height)) # use fps, width and height from orignal video to compile new video 
```

```python
# print landmark data 
# currently saved as one row per label for each frame. 32 rows per frame 
df.tail()
```

```python
# Group landmark data by frame
grouped = df.groupby("frame")
print(grouped.get_group(30))
```

```python
print(grouped.get_group(100))
```

```python
frame_idx = 0 # start at frame zero 

while cap.isOpened(): # checks object cap is open and ready to read frames 
    ret, frame = cap.read() # ret: True = succesffuly read, False = no more frames or an error; # frame = numpy array of that frame's data 
    if not ret: # break if video has ended or error reading frames 
        print('end video or error reading frames')
        break
    
    # video hasn't ended and no error reading frames 
    # Get landmark data for this frame (if available)
    print(frame_idx)
    if frame_idx in grouped.groups:
        group = grouped.get_group(frame_idx)

        # Skip invalid frames
        if len(group) == 1 and group.iloc[0]['vis'] == float('inf'): # only one row per frame and vis = 'inf'
            frame_idx += 1 # move to next frame 
            out.write(frame)
            continue

        # if landmark is valid: go through each row in this frame group and save x y z and visibility  
        landmarks = []

        for _, row in group.iterrows():
            landmark = landmark_pb2.NormalizedLandmark(
                x=row["X"], y=row["Y"], z=row["Z"], visibility=row["vis"]
            )
            landmarks.append(landmark)

        if len(landmarks) == 33:
            landmark_list = landmark_pb2.NormalizedLandmarkList(landmark=landmarks)

            mp_drawing.draw_landmarks(
                frame,
                landmark_list,
                mp_pose.POSE_CONNECTIONS,
                landmark_drawing_spec=mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2),
                connection_drawing_spec=mp_drawing.DrawingSpec(color=(0, 0, 255), thickness=2)
            )
            
            
    # Write the frame (with or without landmarks)
    out.write(frame)
    frame_idx += 1

cap.release()
out.release()
cv2.destroyAllWindows()
```
