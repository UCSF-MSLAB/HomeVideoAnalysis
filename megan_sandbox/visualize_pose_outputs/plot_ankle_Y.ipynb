```python
## Import libraries 
import matplotlib.pyplot as plt 
import pandas as pd 
import os 
```

```python
# input folder 
folder_path = r"C:\Users\mmccu\Box\MM_Personal\5_Projects\BoveLab\3_Data_and_Code\gait_hc_outputs\MM_HC\2024-07-30\000_raw_pose_data"

# mp pose 
input_file_mp_pose = r"MM_HC_17ft_gait_vertical_right_mediapipe.csv"
df_mp = pd.read_csv(os.path.join(folder_path, input_file_mp_pose))

# mp world 
input_file_mp_world = r"MM_HC_17ft_gait_vertical_right_mediapipe_world.csv"
df_mp_world = pd.read_csv(os.path.join(folder_path, input_file_mp_world))

# yolo pixels 
input_file_yolo = r"MM_HC_17ft_gait_vertical_right_yolo.csv"
df_yolo = pd.read_csv(os.path.join(folder_path, input_file_yolo))
```
